```python
import cfbd
import numpy as np
import pandas as pd
import requests
import json
from API_Key import key
import matplotlib.pyplot as plt
import seaborn as sns
from factor_analyzer import FactorAnalyzer

import xgboost as xgb
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold, train_test_split
import shap
```

```python
configuration = cfbd.Configuration()
configuration.api_key['Authorization'] = key
configuration.api_key_prefix['Authorization'] = 'Bearer'
```

```python
def get_basic_statdata(year):
    api_instance = cfbd.GamesApi(cfbd.ApiClient(configuration))
    games = api_instance.get_games(year=year)
    games[0]


    year_games = pd.DataFrame().from_records([
        g.to_dict()
        for g in games
    ]
    )

    year_games['winner']=np.where(year_games.home_points > year_games.away_points, year_games.home_team, year_games.away_team)
    year_games['diff'] = np.abs(year_games.home_points - year_games.away_points)


    # pd.options.display.max_columns = 100
    year_games = year_games[['id','home_team','away_team','home_points','away_points','diff','excitement_index','winner']]

    api_instance = cfbd.StatsApi(cfbd.ApiClient(configuration))
    stats = api_instance.get_advanced_team_game_stats(year=year)

    stat_df = pd.DataFrame().from_records([
        s.to_dict() for s in stats
    ]
    )

    #offense columns
    offense_columns = list()
    offense = pd.json_normalize(stat_df['offense'])

    for i in offense.columns:
         offense_columns.append(i + '_offense')

    offense = pd.json_normalize(stat_df['offense'])
    offense.columns = offense_columns
    offense.columns


    #defense columns
    defense_columns = list()
    defense = pd.json_normalize(stat_df['defense'])

    for i in defense.columns:
         defense_columns.append(i + '_defense')

    defense.columns = defense_columns
    defense.columns

    #combine
    stats = pd.concat([stat_df.drop(['offense','defense'],axis=1),offense,defense],axis=1)
    stats.rename(columns={'game_id':'id'},inplace=True)

    combined = pd.merge(left = year_games,right = stats,on='id')

    combined['win']=np.where(combined.team == combined.winner,1,0)
    combined['adj_diff'] = np.where(combined.win == 1, combined['diff'], -combined['diff'])
    data = combined.drop(['id','home_team','away_team','home_points','away_points','winner','season','week','team','opponent'],axis=1)
    
    
    return combined
```

```python
def get_basic_training_data(start_year, end_year):
    year_data = []
    for year in range(start_year,end_year+1):
        year_data.append(get_basic_statdata(year))
    return pd.concat(year_data,ignore_index=True)

train_ = get_basic_training_data(2014,2018)
test_ = get_basic_training_data(2019,2019)
```

```python
train = train_.drop(['id','home_team','away_team','home_points','away_points','diff','winner','season','week','team','opponent'],axis=1)
test = test_.drop(['id','home_team','away_team','home_points','away_points','diff','winner','season','week','team','opponent'],axis=1)

model = xgb.XGBRegressor()
y_train,X_train = train.iloc[:,-1],train.iloc[:,:-1]
y_test,X_test = test.iloc[:,-1],test.iloc[:,:-1]

model.fit(X_train,y_train)

cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
# evaluate model
scores = cross_val_score(model, X_train, y_train, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)

scores = np.absolute(scores)
print('Mean MAE: %.3f (%.3f)' % (scores.mean(), scores.std()) )
```

```python
shap.initjs()
e = shap.TreeExplainer(model, X_train)
shap_values = e.shap_values(X_train)
shap.summary_plot(shap_values, X_train)
```

```python
def lin_transform(df):   
    data = df.drop(['id','home_team','away_team','home_points','away_points','winner','season','week','team','opponent','diff'],axis=1)

    all_vars = dict()
    for i in data.columns:
        all_vars[i] = None

    srs = pd.DataFrame()
    count = 1
    for var in data.columns:
        combo_var = df.groupby('team').mean()[var]
        terms = []
        solutions = []

        for team in combo_var.keys():
            row = []
            opps = list(df[df['team'] == team]['opponent'])

            for opp in combo_var.keys():
                if opp == team:
                    # coefficient for the team should be 1
                    row.append(1)
                elif opp in opps:
                     # coefficient for opponents should be 1 over the number of opponents
                    row.append(-1.0/len(opps))
                else:
                    # teams not faced get a coefficient of 0
                    row.append(0)

            terms.append(row)

            # average game spread on the other side of the equation
            solutions.append(combo_var[team])

        solutions = np.linalg.solve(terms,solutions)

        if count == 1:
            ratings = list(zip( combo_var.keys(), solutions ))
            srs = pd.DataFrame(ratings, columns=['team', var])
        else:
            srs[var] = solutions

        # normalize the data so that the average team has a rating of 0
        mean = srs[var].mean()
        srs[var] = srs[var] - mean

        count += 1

    return srs
```

```python
test_data = lin_transform(test_)
```

```python
def get_spread(teams):
    team1 = teams[0]
    team2 = teams[1]
    game = test_data[test_data['team'].isin([team1,team2])]
    return game[game['team'] == team1]['test'].values[0] - game[game['team'] == team2]['test'].values[0]

get_spread(('Georgia','LSU'))
```

```python
test_['spread'] = [-get_spread(i) for i in list(zip(test_['team'],test_['opponent']))]
```

```python
data_with_spreads = pd.merge(left = test_,right = betting_spread,right_on=['id','home_team','away_team'],left_on = ['id','team','opponent'])
# data_with_spreads[data_with_spreads['team'] == 'Georgia']
data_with_spreads.rename(columns= {'spread_x':'pred_spread','spread_y':'act_spread'},inplace=True)
data_with_spreads
data_with_spreads[['team','opponent','adj_diff','test','pred_spread','act_spread']]
data_with_spreads['act_spread'].astype(float)
data_with_spreads['pred_winner'] = np.where(data_with_spreads.pred_spread < data_with_spreads['act_spread'].astype(float), data_with_spreads.team, data_with_spreads.opponent)
data_with_spreads['act_winner'] = np.where(-data_with_spreads.adj_diff < data_with_spreads['act_spread'].astype(float), data_with_spreads.team, data_with_spreads.opponent)
data_with_spreads['correct'] = np.where(data_with_spreads.pred_winner == data_with_spreads.act_winner,1,0)
model_acc = data_with_spreads['correct'].sum()/len(data_with_spreads)
print(f'Accuracy {model_acc : .3f}')
# data_with_spreads.sort_values(by='excitement_index',ascending=False)
```
